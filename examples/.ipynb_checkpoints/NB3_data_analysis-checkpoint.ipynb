## Clustering analysis for crime data (labeled)

### By types of institution (private or public)

```python
%%R

for (i in c(1:9,99)){
  cat(sprintf('sector ID %d : %s has %d institutions\n',i,data$Sector_desc[data$sector_cd == i][1],sum(data$sector_cd == i)))
}
```

The type $99$ only has $1$ institution, so we are focusing on the type $1$ to $9$. We devide those institutions by three types, $4$-year colleges, $2$-year colleges and less-than $2$-year colleges.

Clustering for 4-year colleges.

Year 2010

```python
%%R

#taking out of population issue
for (i in names(data10)[13:22]){
  data10[[i]] <- data10[[i]]/data10$Total
}
#normalization
for (i in names(data10)[13:22]){
  data10[[i]] <- (data10[[i]]-mean(data10[[i]]))/sd(data10[[i]])
}
```

Clustering for all 9 variables.

```python
%%R -r 86 -w 800 -h 400

colors = c('#FF000010', '#00FF0010', '#0066FF10')
r <- 0
for (i in names(data10)[13:22]){
  r <- r + data10[[i]]
}
data10$r <- r
xpoints <- c(data10$r[data10$sector_cd == 1],data10$r[data10$sector_cd == 2],data10$r[data10$sector_cd == 3])
ypoints <- c(rep(1,sum(data10$sector_cd == 1)),rep(0,sum(data10$sector_cd == 2)),rep(-1,sum(data10$sector_cd == 3)))
sector <- c(data10$sector_cd[data10$sector_cd == 1],data10$sector_cd[data10$sector_cd == 2],data10$sector_cd[data10$sector_cd == 3])
plot(xpoints,ypoints,ylim = c(-1.5,3),xlim=c(-1,5), col=colors[sector],pch=16,xlab='crime index',ylab ='',main='Clustering for 4-year Colleges')
legend('topright', legend=c('public 2010','private nonprofit 2010','private profit 2010'), col=c('red','green','blue'), pch=16)
abline(v=mean(data10$r[data10$sector_cd == 1]),col='red')
abline(v=mean(data10$r[data10$sector_cd == 2]),col='green')
abline(v=mean(data10$r[data10$sector_cd == 3]),col='blue')
```

### clustering for two-dimension

we need to decompose our 9 variables as two components

one reasonable way:

r1: Burglary (the most common one)

r2: other types of crime (rare)

year 2010

```python
%%R -r 86 -w 800 -h 400

#plot in two-dimension
#dimension one
names(data10)
r1 <- 0
for (i in names(data10)[c(19)]){
  r1 <- r1 + data10[[i]]
}
#dimension two
r2 <- 0
for (i in names(data10)[c(13:18,20:21)]){
  r2 <- r2 + data10[[i]]
}
#plot
colors = c('#0000FF50','#FF000050')
xpoints <- r1[data10$State == 'CA' & data10$BRANCH == 'Main Campus']
ypoints <- r2[data10$State == 'CA' & data10$BRANCH == 'Main Campus']
plot(xpoints,ypoints, col=colors[1],pch=16,xlab='crime index 1',ylab ='crime index 2',main='Crime in CA colleges in 2010')
```

k-means

```python
%%R -r 86 -w 800 -h 400

#k = 2
points <- cbind(xpoints,ypoints)
fit = kmeans(points, 2)
plot(points,col = c('red', 'blue')[fit$cluster],pch=16,xlab='crime index 1',ylab ='crime index 2',main='Crime in CA colleges in 2010')
```

So, we can ...

```python
%%R -r 86 -w 800 -h 400

#k = 3
fit = kmeans(points, 3)
plot(points,col = c('red', 'blue','green')[fit$cluster],pch=16,xlab='crime index 1',ylab ='crime index 2',main='Crime in CA colleges in 2010')
```

```python
%%R

print(data10$INSTNM[data10$State == 'CA' & data10$BRANCH == 'Main Campus'][fit$cluster==2])
```

These colleges have fewer burglary crime and more other types of crime.

```python
%%R

print(data10$INSTNM[data10$State == 'CA' & data10$BRANCH == 'Main Campus'][fit$cluster==3])
```

The k-means clustering method fits $4$ groups. One for ...
